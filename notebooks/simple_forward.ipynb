```python
%matplotlib inline

import matplotlib.pyplot as plt
import xarray as xr

from fair_spice.fair.models import forward_fair
from fair.RCPs import rcp6
```

## Basic fair model run

Run the FAIR model in forward mode and plot the radiative forcings by source.

```python
run = forward_fair({"emissions": rcp6.Emissions.emissions})
run.time.values[:] = rcp6.Emissions.year
```

```python
fig, ax = plt.subplots(figsize=(12,12))

for forcing in run.forcing:
    run.radiative_forcing.sel(forcing=forcing).plot(ax=ax, label=forcing.values)

ax.legend()
ax.set_title("Radiative Forcings in RCP6")
ax.set_ylim(-2, 7)
```
